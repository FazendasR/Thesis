```python
import pickle 

def load_all_pickle_dicts_from_folder(folder_path):
    """
    Loads all pickle files in the given folder and returns a dictionary of dictionaries.

    :param folder_path: Path to the folder containing .pkl files.
    :return: A dictionary where keys are filenames (without .pkl) and values are the loaded dictionaries.
    """
    all_dicts = {}

    for filename in os.listdir(folder_path):
        if filename.endswith('.pkl'):
            file_path = os.path.join(folder_path, filename)
            with open(file_path, 'rb') as f:
                dict_key = os.path.splitext(filename)[0]
                all_dicts[dict_key] = pickle.load(f)

    return all_dicts
```

```python
cleaned_data = load_all_pickle_dicts_from_folder(dc.HP_PATH_CLEANED_DOCS_DICTS)
```
